# IMPORTING LIBRARIES

```python
# IMPORTING NUMPY
import numpy as np
# IMPORTING PANDAS
import pandas as pd
# IMPORTING TENSORFLOW
import tensorflow as tf
# IMPORTING MATPLOTLIB
import matplotlib.pyplot as plt
# IMPORTING JSON
import json
```

# LOAD JSON FILE

```python
# OPEN JSON FILE
JSON_DATA=open('Intent.json','r')
# CONVERT DATA INTO DICTIONARY FORMAT
data=json.load(JSON_DATA)
# GET INTENT DATA FROM THE DICTIONARY
intents_list= data['intents']
# EMPTY LIST FOR TAG
tags=[]
# EMPTY LIST FOR INPUT 
inputs=[]
for data in intents_list:
  patterns=data['text']
  for p in patterns:
    tags.append(data['intent'])
    inputs.append(p)
```

```python
# TAGS
tags
```

```python
# INPUTS
inputs
```

```python
# CONVERT DATA INTO DATAFRAME
dataset=pd.DataFrame(
    {
        'inputs':inputs,
     'tags':tags
    }
)
# VIEW DATASET
dataset
```

# TEXT PROCESSING

```python
import re
corpus=[]
for indx in range(0,len(dataset['inputs'])):
    temp=re.sub('[^A-Za-z]',' ',dataset['inputs'][indx])
    temp=re.sub(' +',' ',temp)
    corpus.append(temp)
```

```python
corpus
```

```python
tokenizer=tf.keras.preprocessing.text.Tokenizer(num_words=2000)
tokenizer.fit_on_texts(corpus)
X=tokenizer.texts_to_sequences(corpus)
X=tf.keras.preprocessing.sequence.pad_sequences(X)
Y=dataset.iloc[:,-1].values
```

```python
print(X)
```

```python
len(X[0])
```

```python
print(Y)
```

```python
from sklearn.preprocessing import LabelEncoder
encodeY=LabelEncoder()
Y_encoded=encodeY.fit_transform(Y)
Y_encoded
```

```python

Y=Y.reshape(len(Y),1)
print(Y)
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
CT=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[0])],remainder='passthrough')
Y=CT.fit_transform(Y).toarray()
```

```python
print(Y)
```

```python
Y.shape
```

```python
tokenizer.word_index
```

```python
MODEL=tf.keras.models.Sequential()
MODEL.add(tf.keras.layers.Dense(units=30,activation='relu'))
MODEL.add(tf.keras.layers.Dense(units=60,activation='relu'))
MODEL.add(tf.keras.layers.Dense(units=90,activation='relu'))
MODEL.add(tf.keras.layers.Dense(units=100,activation='relu'))
MODEL.add(tf.keras.layers.Dense(units=150,activation='relu'))
MODEL.add(tf.keras.layers.Dense(units=31,activation='softmax'))
MODEL.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
MODEL.fit(X,Y,batch_size=10,epochs=400)
```

```python
input_text=np.array(tokenizer.texts_to_sequences(['Hi there']))
padding_size=9-len(input_text[0])
input_text=np.pad(input_text[0],(padding_size,0),'constant',constant_values=(0,))
input_text=[list(input_text)]
input_text=np.array(input_text)
input_text
```

```python
import random


def CHATBOT(input):
    input_text = np.array(tokenizer.texts_to_sequences([input]))
    padding_size = 9-len(input_text[0])
    input_text = np.pad(input_text[0], (padding_size, 0),
                    'constant', constant_values=(0,))
    input_text = [list(input_text)]
    input_text = np.array(input_text)
    
    P=MODEL.predict(input_text)
    intent=encodeY.inverse_transform([np.argmax(P)])
    for data in intents_list:
        if data['intent']==intent:
            res=data['responses']
            break
    res_indx=random.randint(0,len(res)-1)
    print("CHATBOT : "+res[res_indx])
```

```python
while(True):
    user_input=input("IN :")
    if user_input=='0':
        break
    else:
        CHATBOT(user_input)
```
